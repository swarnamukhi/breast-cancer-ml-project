```python
df= pd.read_csv("/content/wdbc.data", names=columns)
```

```python
df.drop("id",axis=1,inplace=True)
```

```python
x=df.drop(columns=["diagnosis"])
y=df["diagnosis"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
x_scaled=scaler.fit_transform(x_train)
x_test_scaled=scaler.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logistic=LogisticRegression()
```

```python
logistic.fit(x_scaled,y_train)
```

```python
logistic.classes_
```

```python
y_predict=logistic.predict_proba(x_test_scaled)
```

```python
y_predict
```

```python
y_predict[:,-1]
```

```python
type(y_predict)
```

```python
threshold=0.1
```

```python
y_pred_custom = (y_predict[:, -1] >= threshold).astype(int)
```

```python
y_pred_custom
```

```python
from sklearn.metrics import *
```

```python
Y_pred = pd.DataFrame(y_pred_custom, columns=['diagnosis'])
```

```python
Y_pred['diagnosis']=Y_pred['diagnosis'].map({1:'M',0:'B'})
```

```python
confusion_matrix(y_test,Y_pred)
```

In this step, we explored adjusting the classification threshold in the Logistic Regression model to reduce false negatives (FN), which is critical in medical datasets like breast cancer detection.

By default, the model uses a threshold of 0.5 to convert predicted probabilities into class labels. To gain more control over predictions, we used `predict_proba()` to extract the probability of the positive class (Malignant) and applied custom thresholds (0.5 and 0.1).

The goal was to reduce false negatives, as missing a malignant case is more serious than a false alarm.

However, after evaluation using the confusion matrix:
- At threshold = 0.5 → FN = 1
- At threshold = 0.1 → FN = 1

We observed that lowering the threshold did not reduce false negatives. Instead, it increased false positives.

This indicates that the misclassified malignant case had a very low predicted probability, meaning the model was highly confident but incorrect. In such cases, threshold tuning alone cannot fix the issue.

Conclusion:
Threshold adjustment is effective only for borderline cases (probabilities near the decision boundary). Since our false negatives were not borderline, this method did not improve model performance. Therefore, we retain the default threshold (0.5) and explore other techniques like class weighting to further reduce false negatives.
